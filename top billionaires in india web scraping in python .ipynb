```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_Indians_by_net_worth'
response = requests.get(url)
```

```python

soup = BeautifulSoup(response.text, 'html')
```

```python
print(soup)
```

```python
soup.find('table')
```

```python
table = soup.find('table',class_ ="wikitable sortable" )
```

```python
print(table)
```

```python
top_list = soup.find_all('th')
print(top_list)
```

```python
top_list_title = [title.text.strip() for title in top_list]
print(top_list_title)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = top_list_title)
print(df)
```

```python
col_data =table = soup.find_all('tr')
```

```python
for row in col_data[1:33]:
  row_data = row.find_all('td')
  individual_row_data = [data.text.strip() for data in row_data]
  print(individual_row_data)
```

```python
df.to_csv(r'F:\excel files\toplists')
```
